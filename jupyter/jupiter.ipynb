# 2. DataFrames

```python
dc1 = {
    'ID' : [1,2,3],
    'Name' : ['A', 'B', 'C'],
    'Age' : [21,22,23]
    }

dc2 = {
    'ID' : [3, 5, 4],
    'Name' : ['C', 'D', 'E'],
    'Age' : [23, 24, 25]
}

df1 = pd.DataFrame(dc1)
df2 = pd.DataFrame(dc2)
print(df1)
print(df2)
```

# 3. Concatenating DataFrames

```python
result = pd.concat([df1, df2], ignore_index=True)
print(result)
```

# 4. Dropping Duplicates

```python
cleanResult = result.drop_duplicates()
print(cleanResult)
```

# 5. Sorting of values

```python
cleanResult = cleanResult.sort_values(by= 'ID')
print(cleanResult)
```
