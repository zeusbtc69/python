```python
placement = placement.drop(columns=['sl_no'])
```

```python
cols = [
    'gender',
    'hsc_b',
    'hsc_s',
    'degree_t',
    'workex',
    'specialisation',
    'status'
]

for c in cols:
    placement[c] = le.fit_transform(placement[c])
```

```python
X = placement.iloc[:,0:-1].values
Y = placement.iloc[:,-1].values
```

# 1. Linear Regression

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=4)
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
```

```python
X_train
sc
```

```python
X_test = sc.transform(X_test)
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, Y_train)
```

```python
yPred = reg.predict(X_test)
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
MSE = metrics.mean_squared_error(Y_test, yPred)
MSE
```

```python
RMSE = np.sqrt(MSE)
RMSE
```

```python
R2 = metrics.r2_score(Y_test, yPred)
R2
```

# 2. Logistic Regression

```python
X = placement.drop(['status', 'salary'], axis = 1).values
X
```

```python
Y = placement['status'].values
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=46)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()

logreg.fit(X_train,Y_train)
```

```python
yPred = logreg.predict(X_test)
```

```python
yPred
```

```python
accuracy_score(Y_test, yPred)
```

```python
metrics.roc_auc_score(Y_test, yPred)
```

```python
data = metrics.confusion_matrix(Y_test, yPred)
```

```python
tn, fp, fn , tp = metrics.confusion_matrix(Y_test, yPred).ravel()
```

```python
ax = sns.heatmap(data, annot=True)
ax.invert_yaxis()
ax.invert_xaxis()
```

```python
r = tp/(tp+fn)
```

```python
p=(tp/(tp+fp))
```

```python
r #recall
```

```python
p #precision
```

```python
print(metrics.classification_report(Y_test, yPred))
```
