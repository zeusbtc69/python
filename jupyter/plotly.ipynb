```python
from sklearn.datasets import make_regression
import pandas as pd 
import numpy as np

import plotly.express as px
import plotly.graph_objects as go

import sklearn.metrics as mx 
```

```python
x, y = make_regression(n_samples=100, n_features=2, n_informative=2, n_targets=1, noise=50)
```

```python
df = pd.DataFrame({'feature1' : x[:,0], 'feature2' : x[:,1], 'target' : y})
```

```python
df.head()
```

```python
fig = px.scatter_3d(df, x='feature1', y='feature2', z='target')
fig.show()
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.4, random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(xTrain, yTrain)
```

```python
yPred = lr.predict(xTest)
```

```python
print('MAE', mx.mean_absolute_error(yTest, yPred))
print('MSE', mx.mean_squared_error(yTest, yPred))
print('R2 SCORE', mx.r2_score(yTest, yPred))
```

```python
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

xGrid, yGrid = np.meshgrid(y, x)

# build feature matrix with shape (n_samples, n_features)
final = np.column_stack((xGrid.ravel(), yGrid.ravel()))
z_final = lr.predict(final).reshape(10,10)
z = z_final
```

```python
fig = px.scatter_3d(df, x='feature1', y='feature2', z='target')
fig.add_trace(go.Surface(x = x, y = y, z = z))

fig.show()
```
